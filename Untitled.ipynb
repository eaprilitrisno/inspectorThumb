```python
import time
from selenium import webdriver
from selenium.common.exceptions import NoSuchElementException

options = webdriver.ChromeOptions()
# if headless == True:
#     options.add_argument("headless")
options.add_argument("user-data-dir='./data'") #Your google chrome data
driver = webdriver.Chrome('./chromedriver',chrome_options=options)
driver.implicitly_wait(10)
```

```python
def ikuti():
    try:
        a = driver.find_element_by_xpath("//button[text()='ikuti']")
        a.click()
        return 1

    except NoSuchElementException as e:
#         print e
        return 0

def unikuti():
    try:
        a = driver.find_element_by_xpath("//button[text()='Mengikuti']")
        a.click()
        return 1
        
    except NoSuchElementException as e:
#         print e
        return 0

def get_all_url():
    x = driver.find_elements_by_xpath('//a')
    for s in x:
        try_url[s.get_attribute('href')]=s.text

def write_f(fname,text):
    file = open(fname,"a") 
    file.write(text+'\n') 
    file.close() 
    
def read_f(fname):
    file = open(fname,"r")
    tmp = file.read().split('\n')
    file.close()
    return tmp

def catat_toko():
    try:
        a = driver.find_element_by_class_name('product-page-seller-info__view-shop')
        return a.get_attribute('href')
    except NoSuchElementException as e:
        return 'kosong'

# print catat_toko()
```

```python
data = read_f('skiped_url.txt')
skiped_url={}
for row in data:
    skiped_url[row]=''
loop = 15

url = 'https://shopee.co.id/search?keyword=tshirt&page=#number#&sortBy=sales'

for i in range(0,loop):
    target = url.replace('#number#',str(i))
    driver.get(target)
    get_all_url()
    for curr_url,curr_text in try_url.items():
        if not curr_url in skiped_url:
            print curr_url
            driver.get(curr_url)
            time.sleep(3)
            resp = ikuti()
            if resp == 1:
                txt = catat_toko()
                if not txt in toko_url:
                    write_f('toko_online.txt',txt)
                    toko_url[txt]=''
            time.sleep(3)
            skiped_url[curr_url]=curr_text
            write_f('skiped_url.txt',curr_url)
```
